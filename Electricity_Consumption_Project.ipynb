```python
import pandas as pd
sales = pd.read_csv("/Users/rohitgujjar/Downloads/sales.csv")
sales.head()
```

```python
sales.info()
```

```python
sales.isnull().sum()
```

```python
sales.drop_duplicates(inplace=True)
```

```python
# check missing values .again
sales.isnull().sum()
```

```python
sales['Variable observation metadata scalingFactor'] =\
sales['Variable observation metadata scalingFactor'].fillna(1)
```

```python
sales.rename(columns={
    'Entity properties name': 'Country',
    'Variable observation date': 'Date',
    'Variable observation value': 'Value',
    'Variable observation metadata unit': 'Unit',
    'Variable properties name': 'VariableName'
}, inplace=True)
```

```python
sales['Date'] = pd.to_datetime(sales['Date'])
```

```python
sales['Value'] = pd.to_numeric(sales['Value'])
```

```python
sales.info()
```

```python
sales.head()
```

```python
# Extract year from Date
sales['year'] = sales['Date'].dt.year
```

```python
# calculate year to year change
sales['YoY_Change'] = sales['Value'].pct_change()*100
```

```python
# calculate 3 year moving average
sales['MA_3']= sales['Value'].rolling(window=3).mean()
```

```python
sales.head()
```

```python
sales['Year'] = sales['Date'].dt.year
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid')
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales['Year'], sales['Value'])
plt.title("Annual Electricity Consumption in the USA")
plt.xlabel("Year")
plt.ylabel("Electricity Consumption (kWh)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales['Year'], sales['YoY_Change'], color='orange')
plt.title("Year-over-Year % Change in Electricity Consumption")
plt.xlabel("Year")
plt.ylabel("Percent Change (%)")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(sales['Year'], sales['Value'], label='Annual Consumption')
plt.plot(sales['Year'], sales['MA_3'], label='3-Year Moving Average', linestyle='--', color='red')
plt.title("Electricity Consumption with 3-Year Moving Average")
plt.xlabel("Year")
plt.ylabel("Electricity Consumption (kWh)")
plt.legend()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10,6))
sns.histplot(sales['Value'], kde=True, bins=15, color='green')
plt.title("Distribution of Annual Electricity Consumption")
plt.xlabel("Electricity Consumption (kWh)")
plt.ylabel("Frequency")
plt.show()
```

```python
# Pivot table
pivot = sales.pivot_table(index='VariableName', columns='Year', values='Value', aggfunc='sum')

plt.figure(figsize=(12,4))
sns.heatmap(pivot, annot=True, fmt=".0f", cmap="YlGnBu")
plt.title("Heatmap of Electricity Consumption over Years")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top5 = sales.nlargest(5, 'Value')

plt.figure(figsize=(10,6))
sns.barplot(x='Year', y='Value', data=top5, color="blue")  # Use color instead of palette
plt.title("Top 5 Years with Highest Electricity Consumption")
plt.ylabel("Electricity Consumption (kWh)")
plt.show()
```
